```python
import csv
import matplotlib.pyplot as plt
import os
def load_fitness(filename) :
    with open(filename) as file :
        csv_data = csv.reader(file,delimiter=',')
        fitnesses = []
        for row in csv_data :
            fitnesses.append(float(row[2]))
        return fitnesses
def load_final_pos(filename) :
    with open(filename) as file :
        csv_data = csv.reader(file,delimiter=',')
        x = []
        y = []
        for row in csv_data :
            x.append(float(row[1]))
            y.append(float(row[2]))
        return x, y
def best_fit(fitnesses_vec) :
    best_fits = []
    for fits in fitnesses_vec :
        best_fits.append(max(fits[200:]))
    return best_fits
def load_eval_time(filename) :
    with open(filename) as file :
        csv_data = csv.reader(file,delimiter=',')
        durations = []
        eval_nb = []
        for row in csv_data :
            durations.append(float(row[1])/1000000000.)
            eval_nb.append(int(row[0]))
    return eval_nb, durations
```
