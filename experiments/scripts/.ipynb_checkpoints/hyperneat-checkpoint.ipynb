```python
import csv
import matplotlib.pyplot as plt
import os
def load_fitness(filename) :
    with open(filename) as file :
        print(filename)
        csv_data = csv.reader(file,delimiter=',')

        fitnesses = []
        for row in csv_data :
            fitnesses.append(float(row[406]))
        return fitnesses
def load_final_pos(filename,pop,gen) :
    with open(filename) as file :
        csv_data = csv.reader(file,delimiter=',')
        x = []
        y = []
        for row in csv_data :
            if(float(row[1]) == gen) :
                i = 0
                j = 2
                while(i < pop - 1):
                    if(j >= len(row) -1) :
                        break
                    if(float(row[j]) == i and float(row[j+1]) != i+1) :
                        x.append(float(row[j+1]))
                        y.append(float(row[j+2]))
                        j+=3
                    
                    j+=1
                    i+=1
                break
        return x, y
```

```python
hn_hard_maze_t1 = "/home/le_goff/are-logs/hyperneat_hard_maze_t1/"
for folder in os.listdir(hn_hard_maze_t1) :

    fit = load_fitness(hn_hard_maze_t1 + folder + "/fitnesses_hn.csv")
    fig, ax = plt.subplots()
    plt.plot(fit)
    plt.show()
```
