```python
# Import libraries
import csv
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as lin
import os
import pandas as pd
import seaborn as sns
import warnings
import seaborn as sns
import statistics
```

```python
def load_ctrl_genome(filename):
    with open(filename) as file:
        i = 0
        nbr_weights = 0
        nbr_biases = 0
        params = []
        for line in file.readlines():
            if(i == 0):
                nbr_weights = int(line)
                i+=1
            elif(i == 1):
                nbr_biases  = int(line)
                i+=1
            elif(i == 2):
                params.append(float(line))
        return nbr_weights, nbr_biases, params
```

```python
exp_folder = "/home/leni/are-logs/tri_nipes_multi/"

distances = []
for folder in os.listdir(exp_folder):
    params = []
    for file in os.listdir(exp_folder + "/" + folder):
        if(file.split("_")[0] == "genome"):
            w,b,p = load_ctrl_genome(exp_folder + "/" + folder + "/" + file)
            params.append(p)
    distances += [[i,lin.norm(np.array(params[i-1])-np.array(params[i]))] for i in range(1,len(params))]
    
data = pd.DataFrame(data=distances,columns=["Eval","distance"]) 
```

```python
ax = sns.lineplot(data=data,x="Eval",y="distance")
```

```python
ax = sns.lineplot(data=data,x="Eval",y="distance")
ax.set_xlim(0,100)
```
