```python
# Import libraries
import csv
import matplotlib.pyplot as plt
import numpy as np
import imageio
import moviepy.editor as mp
import os
```

```python
# Set filepath
repository='/home/ebb505/ARE_Project/are-logs/'
replicate='morphology4_3_13-58-4-4592/'
filename='morphdesc.csv'
filepath=repository+replicate+filename
print('Filepath: ' + filepath)
```

```python
# Import data from csv
my_list = []
with open(filepath, 'r') as csvfile:
    data = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in data:
        my_list.append(row[:-1]) # Last element is empty
```

```python
# Experiment values
population = 40
evaluations = len(my_list)
featureSize = len(my_list[0])
print("Number of evaluations: ", evaluations, "Features size: ", featureSize)
```

```python
xAxis = 1
yAxis = 3
xLabel = "Width"
yLabel = "Height"
```

```python
# Plot stuff 
dat = np.array([[float(y) for y in x] for x in my_list]) # Convert to array
images = []
for i in range(population,evaluations,population):
    plt.figure(i)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)

    #plt.scatter(dat[:i,xAxis],dat[:i,yAxis],marker='D')
    #temp=dat[:i,:]

    #plt.scatter(temp[temp[:i,6] > 0,xAxis],temp[temp[:i,6] > 0,yAxis],marker='x')
    #plt.scatter(temp[temp[:,6] == 0,xAxis],temp[temp[:i,6] == 0,yAxis],marker='D'

    # For legend
    plt.scatter([],[],marker='x',c=[1,0,0,1])
    plt.scatter([],[],marker='+',c=[0,0,1,1])

    temp=dat[:i,:]
    length=sum(temp[:i,6] == 0)
    red = []
    red = np.ones(length)
    list(red)
    red.tolist()
    green = []
    green = np.zeros(length)
    list(green)
    green.tolist()
    blue = []
    blue = np.zeros(length)
    list(blue)
    blue.tolist()
    transparecy = []
    #transparecy = np.linspace(0,1,length)
    transparecy = np.geomspace(0.01,1,length)
    list(transparecy)
    transparecy.tolist()
    rgba = []
    for k in range(length):
        rgba+=[[red[k],green[k],blue[k],transparecy[k]]]
    plt.scatter(temp[temp[:,6] == 0,xAxis],temp[temp[:i,6] == 0,yAxis],marker='x',c=rgba)
    length=sum(temp[:i,6] > 0)
    red = []
    red = np.zeros(length)
    list(red)
    red.tolist()
    green = []
    green = np.zeros(length)
    list(green)
    green.tolist()
    blue = []
    blue = np.ones(length)
    list(blue)
    blue.tolist()
    transparecy = []
    #transparecy = np.linspace(0,1,length)
    transparecy = np.geomspace(0.01,1,length)
    list(transparecy)
    transparecy.tolist()
    rgba = []
    for k in range(length):
        rgba+=[[red[k],green[k],blue[k],transparecy[k]]]
    plt.scatter(temp[temp[:,6] > 0,xAxis],temp[temp[:i,6] > 0,yAxis],marker='+',c=rgba)

    plt.legend(['Non-Manufacturable robots','Manufacturable robots'])
    axes = plt.gca()
    axes.set_xlim([0,1.0])
    axes.set_ylim([0,1.0])
    axes = plt.gca()
    title = "No manufacturability - Gen: " + str(i/50)
    plt.title(title)
    #plt.show()
    frame = xLabel + yLabel + str(i) + ".png"
    plt.savefig(repository + replicate + frame)
    images.append(imageio.imread(repository + replicate + frame))
    print("Gen:" + str(i/50))
```

```python
# Save gif and video
gifName = xLabel + yLabel + ".gif"
imageio.mimsave(repository + replicate + gifName, images, duration = 0.25)
clip = mp.VideoFileClip(repository + replicate + gifName)
videoName = xLabel + yLabel + ".mp4"
clip.write_videofile(repository + replicate + videoName)
```
