```python
import csv
import matplotlib.pyplot as plt
import os
def load_fitness(filename) :
    with open(filename) as file :
        print(filename)
        csv_data = csv.reader(file,delimiter=',')

        fitnesses = []
        for row in csv_data :
            fitnesses.append(float(row[406]))
        return fitnesses
```

```python
hn_hard_maze_t1 = "/home/leni/are-logs/hyperneat_hard_maze_t1/"
for folder in os.listdir(hn_hard_maze_t1) :

    fit = load_fitness(hn_hard_maze_t1 + folder + "/fitnesses_hn.csv")
    fig, ax = plt.subplots()
    plt.plot(fit)
    plt.show()
```
